Podemos observar que la fila de salary_range esta muy vacía y que vamos a tener que hacer algo con esa información que falta.

#### Recuperar valores de salary_range
Poner el mismo salary range en caso de que sea el mismo trabajo y de que
su atributo salary_range sea nulo

```python
# Comprobamos si existen trabajos iguales comparando el titulo de la oferta de trabajo
dict = {}
for element in data['title']:
    if not element in dict:
        dict[element] = 1
    else:
        dict[element] += 1
dict
```

```python
# creamos nuevo diccionario donde KEY=title, VALUE=salary_range
jobs_dict = {}

# Guardar filas con salary-range

# con <element> iteramos todos los title del dataset (sin repeticiones)
for element in dict:
    # Obtenemos todas las filas con title = element
    row = data.loc[data['title'] == element]
    min_aux = []
    max_aux = []
    if element not in dict:
        jobs_dict[element] = "0"
    
    # con <num> iteramos todas las columnas(salary_range) de cada fila de row 
    for num in row['salary_range']:
        # si no existe un rango, lo ignora
        try:
            if math.isnan(float(num)) == True:
                pass
        # si existe un rango lo añade a las listas de rangos
        except:
            intervals = num.split('-')
            try:
                min_aux.append(int(intervals[0]))
                max_aux.append(int(intervals[1]))
            except:
                pass
    try:
        # obtiene la media de rangos minimos y maximos para asignarla al trabajo
        min_ = str(int(np.mean(min_aux)))
        max_ = str(int(np.mean(max_aux)))
        jobs_dict[element] = min_ + '-' + max_
    except:
        pass

jobs_dict
```

```python
for element in jobs_dict:
    rows = data.loc[data['title'] == element]
    indices = rows.index
    for j in indices:
        try:
            if np.isnan(data['salary_range'][j]) == True:
                data['salary_range'][j] = jobs_dict[element]
        # si entra aqui tiene un salary_range
        except:
            pass
```

```python
# comprovamos cuantos trabajos existen ahora con salary_range
second_count = 0
for element in data['salary_range'].isnull():
    if element != True:
        second_count+=1
print('Antes -->', first_count, '| Despues -->', second_count)
print('Hemos conseguido generar nuevos valores de salary_range para', second_count - first_count, 'filas')
```

```python
# cantidad de filas con salary_range nulo
nullSalaryRange = len(data[data.salary_range.isnull()])
print("Tamaño actual dataset:", len(data), "\nCantidad de filas con salary_range nulo a eliminar:", nullSalaryRange)
```

```python
# eliminamos filas sin salary_range
delete = data[data.salary_range.isnull()].index
data = data.drop(labels=delete, axis=0)
print("Tamaño del dataset después de tratar salary_range:", len(data))
```

#### Tal y como pasa en salary_range, pasa en otros atributos tambien, entonces creamos una función para simplificar el trabajo

```python
def preprocessing_function (dic, reference, missing):
    # contamos cuantos hay al iniciar
    first_count = 0
    for element in data[missing].isnull():
        if element != True:
            first_count+=1
    first_count

    # creamos nuevo diccionario donde KEY=title, VALUE=missing
    missing_dict = {}
    # con <element> iteramos todos los title del dataset (sin repeticiones)
    for element in dic:
        # Obtenemos todas las filas con title = element
        row = data.loc[data[reference] == element]
        if element not in dic:
            missing_dict[element] = "0"
        
        # con <num> iteramos todas las columnas(missing) de cada fila de row 
        for num in row[missing]:
            try:
                if math.isnan(float(num)) == True:
                    pass
            except:
                missing_dict[element] = num

    # damos el valor a los que sean null
    for element in missing_dict:
        rows = data.loc[data[reference] == element]
        indices = rows.index
        for j in indices:
            try:
                if np.isnan(data[missing][j]) == True:
                    data[missing][j] = missing_dict[element]
            # si entra aqui tiene un missing
            except:
                pass
    
    # vemos nuevos valores hemos creado
    second_count = 0
    for element in data[missing].isnull():
        if element != True:
            second_count += 1
    print('Antes -->', first_count, '| Despues -->', second_count)
    print('Hemos conseguido generar nuevos valores para', second_count - first_count, 'filas')
```

### Como no podemos prepocesar más, y los valores que quedan son categóricos, podem sustituir los valores nulos por una nueva categoría 'non defined'

#### Tenemos que pasar los valores categoricos a numericos, para ello utilizaremos mean encoding

```python
cols = ['title', 'location', 'department', 'company_profile', 'description', 'requirements',
        'employment_type', 'required_experience', 'required_education', 'industry', 'function',
        'salary_range']
target = 'fraudulent'
te = TargetEncoder()
for col in cols:
    # fit the data
    te.fit(X=data[col], y=data[target])
    # transform
    values = te.transform(data[col])
    data[col] = values
```
